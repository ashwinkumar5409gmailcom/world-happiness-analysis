```python
from google.colab import files
import pandas as pd

# Upload a single CSV (e.g., 2021.csv)
uploaded = files.upload()

# Get the filename
for fn in uploaded.keys():
    filename = fn

# Read it
df = pd.read_csv(filename)
df.head()
```

```python
from google.colab import files
uploaded = files.upload()

for fn in uploaded.keys():
    filename = fn

df_2018 = pd.read_csv(filename)
df_2018.head()
```

```python
# Select relevant columns
top_10_2018 = df_2018[['Country or region', 'Score']].sort_values(by='Score', ascending=False).head(10)
top_10_2019 = df[['Country or region', 'Score']].sort_values(by='Score', ascending=False).head(10)

# Rename columns for merging
top_10_2018.columns = ['Country', 'Score_2018']
top_10_2019.columns = ['Country', 'Score_2019']

# Merge
merged = pd.merge(top_10_2018, top_10_2019, on='Country')

# Plot
import matplotlib.pyplot as plt

merged.plot(x='Country', kind='bar', figsize=(12,6), title='Top 10 Happiest Countries: 2018 vs 2019')
plt.ylabel("Happiness Score")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
